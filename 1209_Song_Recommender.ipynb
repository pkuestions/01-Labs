```python
url = "https://www.billboard.com/charts/hot-100/"

def show_df_top_100(url, SaveBool=False):
    
    soup = BeautifulSoup(requests.get(url).content, "html.parser")
    
    song_list = [soup.select(".o-chart-results-list-row-container")[s].h3.text.strip()
         for s in range(len(soup.select(".o-chart-results-list-row-container")))]
    
    artist_list = [soup.select(".o-chart-results-list-row-container")[artist].h3.find_next("span").text.strip()
              for artist in range(len(soup.select(".o-chart-results-list-row-container")))]
    
    df = pd.DataFrame(data={
        "Song": song_list,
        "Artist":artist_list
    })
    
    if SaveBool:
        df.to_csv("Billboard_Top_100.csv", index=False)
        
    return df
```

```python
show_df_top_100(url)
```
